```python
import openai
openai.api_key = "sk-"
```

```python
# list models
models = openai.Model.list()
```

```python
import pandas as pd
df = pd.read_csv("/content/AIBE-XV-ANS - Sheet2.csv")
```

```python
df.head()
```

```python
answers = []
for i in range(0,100):
    response = openai.Completion.create(
        model="text-davinci-003",
        prompt = "Question " + df["Question"][i] + "\n" + "Options " + str(df["Options"][i]),
        temperature=0.7,
        max_tokens=150,
        top_p=0.1,
        frequency_penalty=1.176,
        presence_penalty=0,
        request_timeout=5
    )
    answer = response['choices'][0]['text'].strip()
    answers.append(answer)
```

```python
print(len(answers))
```

```python
df = df.drop(100)
```

```python
# Add the answers to the DataFrame as a new column
df['text-davinci-003_Answer'] = answers
```
