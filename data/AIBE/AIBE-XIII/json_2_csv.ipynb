```python
import json
import re

with open("GPT4-Explnation.txt",'r') as file:
    lines = file.readlines()

Pred_ans = None
Exp_key = None
output = []

for line in lines:

    line = line.strip()
    if not line:
        continue

    if re.match(r"^[abcd]\)", line):
        if Pred_ans is not None and Exp_key is not None:
            output.append({"Pred_ans": Pred_ans, "Explanation": Exp_key})

        Pred_ans = line
        Exp_key = None

    else:
        Exp_key = line if Exp_key is None else Exp_key + " " + line


if Pred_ans is not None and Exp_key is not None:
    output.append({"Pred_ans": Pred_ans, "Explanation": Exp_key})

output_json = json.dumps(output, indent=4)

with open("AIBE-XIII-GPT4-Exp.json", "w") as file:
    file.write(output_json)
```

```python
import json

# Load second file
with open('questions_pretty_with_answers.json') as json_file1:
    data1 = json.load(json_file1)

# Load first file
with open('AIBE-XIII-GPT4-Exp.json') as json_file2:
    data2 = json.load(json_file2)

# Combine dictionaries from both files
combined_data = [dict(**d1, **d2) for d1, d2 in zip(data1, data2)]

# Save combined data to a new file
with open('AIBE-XIII-combined.json', 'w') as outfile:
    json.dump(combined_data, outfile, indent=4)
```
